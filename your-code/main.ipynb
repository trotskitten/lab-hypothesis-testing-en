In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
# Your code here
sample_df = df[df["Salary or Hourly"]=="Hourly"]['Hourly Rate']
control_value = 30
alpha = 0.05 #95% confidence interval
p_value = st.ttest_1samp(sample_df,
                         control_value).pvalue #p value of the two-sided test

h_zero = f"the mean of the sample is equal to {control_value}"
h_one = f"the mean of the sample is not equal to {control_value}"
h_zero_rejected = p_value < alpha
print(p_value)
print(f"H0 rejected: {h_zero_rejected}")
if h_zero_rejected:
    print(h_one)
else:
    print(h_zero)
```

While testing our hypothesis is a great way to gather empirical evidence for accepting or rejecting the hypothesis, another way to gather evidence is by creating a confidence interval. A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval.

To read more about confidence intervals, click [here](https://en.wikipedia.org/wiki/Confidence_interval).


In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. 

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
# Your code here
from scipy.stats import t
degrees_freedom = len(sample_df)-1 #degrees of freedom are equal to number of rows -1
sample_mean = sample_df.mean()
standard_error = sem(sample_df, axis=0, ddof=1, nan_policy='propagate', keepdims=False)

conf_int = list(map(float,t.interval(
    confidence = 0.95, #confidence interval
    df = degrees_freedom,
    loc = sample_mean,
    scale = standard_error
    )))

print(f"95% confidence interval: {conf_int}")
```

# Challenge 4 - Hypothesis Tests of Proportions

Another type of one sample test is a hypothesis test of proportions. In this test, we examine whether the proportion of a group in our sample is significantly different than a fraction. 

You can read more about one sample proportion tests [here](http://sphweb.bumc.bu.edu/otlt/MPH-Modules/BS/SAS/SAS6-CategoricalData/SAS6-CategoricalData2.html).



## Hypothesis Testing for Proportions

Hypothesis testing for proportions involves determining whether the proportion of success in a sample is significantly different from a hypothesized proportion. This is commonly used in scenarios where you're interested in categorical outcomes, such as polling, quality control, etc.

### Key Concepts

- **Proportion**: A proportion represents a part of a whole and is often denoted by \( p \).

- **Null Hypothesis (\(H_0\))**: Assumes that the sample proportion is equal to the hypothesized population proportion (\( p_0 \)).

- **Alternative Hypothesis (\(H_1\))**: Suggests that the sample proportion differs from the hypothesized population proportion.

### Practical Example using SciPy

Let's say we want to test if the proportion of defective items in a batch is different from 5% (0.05):

1. **Hypotheses**:
   - \(H_0\): \( p = 0.05 \)
   - \(H_1\): \( p \neq 0.05 \)

2. **Data**: Suppose we have a sample of 200 items with 15 defective ones.

Here's how to conduct the hypothesis test using `scipy.stats`:



```python
from statsmodels.stats.proportion import proportions_ztest

# Data
n = 200  # Total number of items
successes = 15  # Number of defective items
p_null = 0.05  # Hypothesized population proportion

# Conducting the z-test
stat, p_value = proportions_ztest(count=successes, nobs=n, value=p_null)

# Results
print(f"Statistic: {stat}, P-value: {p_value}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis. The proportion is significantly different from 0.05.")
else:
    print("Fail to reject the null hypothesis. The proportion is not significantly different from 0.05.")
```
